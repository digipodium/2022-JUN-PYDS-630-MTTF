**actual preprocessing for this dataset**
- drop the columns that are not needed
- rename the columns
- rename country with very long name (optional)
- convert all columns to str name
- store the range of years in to a variable
- calculate the total immigration per country and store it in a column
- set country as index column



````python
df.head()
````

````python
cols_to_drop = ['Type','Coverage','AREA','REG','DEV']
df.drop(columns=cols_to_drop, inplace=True)
````

````python
df.head(2)
````

````python
df.rename({
    'OdName': 'Country',
    'AreaName': 'Continent',
    'RegName': 'Region',
    'DevName': 'Status',
}, axis=1, inplace=True)
````

````python
df.head(1)
````

````python
df.columns = df.columns.astype(str) # convert column names to strings (very very specific to this data)
````

````python
years = list(map(str,range(1980,2014))) # list of years
print(years)
````

````python
df['Total'] = df[years].sum(axis=1) # sum of all years
df.head(3)
````

````python
df.set_index('Country', inplace=True)
df.head()
````
