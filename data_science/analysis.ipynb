**actual preprocessing for this dataset**
- drop the columns that are not needed
- rename the columns
- rename country with very long name (optional)
- convert all columns to str name
- store the range of years in to a variable
- calculate the total immigration per country and store it in a column
- set country as index column



````python
df.head()
````

````python
cols_to_drop = ['Type','Coverage','AREA','REG','DEV']
df.drop(columns=cols_to_drop, inplace=True)
````

````python
df.head(2)
````

````python
df.rename({
    'OdName': 'Country',
    'AreaName': 'Continent',
    'RegName': 'Region',
    'DevName': 'Status',
}, axis=1, inplace=True)
````

````python
df.head(1)
````

````python
df.columns = df.columns.astype(str) # convert column names to strings (very very specific to this data)
````

````python
years = list(map(str,range(1980,2014))) # list of years
print(years)
````

````python
df['Total'] = df[years].sum(axis=1) # sum of all years
df.head(3)
````

````python
df.set_index('Country', inplace=True)
df.head()
````

**visualization**

````python
plt.style.use('seaborn')
````

````python
df.loc['Japan',years].plot(kind='line', figsize=(15,6), color='red', linewidth=3,marker='o')
````

````python
df.loc['India',years].plot(kind='line', figsize=(15,6), color='blue', linewidth=3,marker='o')
````

````python
countries = ['Japan','France']
df.loc[countries,years].T.plot(kind='line',
                            figsize=(15,6),
                            color=['red','blue'], 
                            linewidth=3,
                            marker='o',
                            markersize=5,)
````

````python
# grouping data by continent
continent_df = df.groupby('Continent').sum()
````

````python
# filter data by continent
filter_europe = df['Continent'] == 'Europe'
eu_df = df[filter_europe].copy()
````

````python
eu_df.sort_values(by='Total', ascending=False, inplace=True)
eu_df.head()[years].T.plot(kind='line',
                            figsize=(15,6),
                            linewidth=2,
                            marker='o',
                            markersize=5,
                            title='Top countries with immigration in Europe',
                            )
````

````python
eu_df.sort_values(by='Total', ascending=False, inplace=True)
eu_df.head(6)[years].T.plot(kind='line',
                            figsize=(20,6),
                            linewidth=2,
                            marker='o',
                            markersize=5,
                            title='Top countries with immigration in Europe',
                            subplots=True,
                            layout = (2,3)
                            )
plt.show()
````

````python
df.sort_values(by='Total',ascending=False,inplace=True)
df.head(6)[years].T.plot(kind='area',
                            figsize=(20,6),
                            linewidth=2,
                            title='Top countries with immigration in World',
                            subplots=True,
                            layout = (2,3)
                            )
plt.show()
````
