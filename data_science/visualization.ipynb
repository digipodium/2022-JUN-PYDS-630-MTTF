```python
import numpy as np
import matplotlib.pyplot as plt # visualization as image 
import seaborn as sns # statistical data visualization
import plotly.express as px # interactive plotting 
```

```python
print(plt.style.available)
```

```python
plt.style.use('seaborn-paper')
```

```python
data = np.random.randint(0,100,50)
print(data)
```

```python
plt.figure(figsize=(25,5))
plt.plot(data)
```

```python
plt.figure(figsize=(25,5))
plt.plot(data,
         color='red',
         linewidth=3,
         linestyle=':',
         marker='*',
         markersize=10)
```

```python
fig = plt.figure(figsize=(25,5))
plt.plot(data,'g--x')
plt.title("Simple Line Plot demo",fontsize=16,color='red',)
plt.xlabel("X axis wala data")
plt.ylabel("Y axis wala data")
```

```python
fig = plt.figure(figsize=(25,5))
plt.plot(data,'g--x')
names = ['Apple','Banana','Cherry','Mango']
plt.xticks([1,15,30,50], labels=names, rotation=45, color='red', fontsize=16)
plt.show()
```

```python
data1 = np.random.randint(0,100,50)
data2 = np.random.randint(0,100,50)
fig = plt.figure(figsize=(25,5))
plt.plot(data1,'r:',label='data1',)
plt.plot(data2,label='data2')
plt.legend(loc='lower right',fontsize=16)
plt.savefig("sample.jpg",bbox_inches='tight')
```

```python
plt.figure(figsize=(25,10))
plt.subplot(2,1,1)
plt.plot(data1,'r:',label='data1',)
plt.subplot(2,1,2)
plt.plot(data2,label='data2')
plt.legend(loc='lower right',fontsize=16)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(25,10),nrows=2,ncols=1,sharex=True) # 0 and 1 are index of subplot
ax[0].plot(data1,'rx:',label='data1',)
ax[1].plot(data2,'bo--',label='data2')
ax[0].legend(loc='lower right',fontsize=16)
ax[1].legend(loc='lower right',fontsize=16)
ax[0].grid(True)
ax[1].grid(True)
plt.show()
```
