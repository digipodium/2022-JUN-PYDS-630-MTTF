```python
# saving this into a file
data = ""
folder = 'C:/windows/system32/'
files  = os.listdir(folder)
for item in files:
    fullpath = os.path.join(folder,item)
    if os.path.isfile(fullpath):
        sz = os.path.getsize(fullpath)
        out = f'{item.ljust(70)} | {sz/1024 :.1f} kb'
        print(out)
        data += out + '\n'
```
